```python
import os
from baselines import results_plotter
%matplotlib inline
```

```python
algs = ["a2c", "acer", "acktr", "deepq", "ppo2"]

envs = ["BeamRider", "Breakout", "Seaquest", "Qbert", "Pong", "Enduro", "SpaceInvaders"]
env_extension = "NoFrameskip-v4"
envs = [env+env_extension for env in envs]

n_steps = ["1e7"]

iters = range(0,10)
iters = [0]
```

```python
for env in envs:
    for alg in algs: 
        for steps in n_steps: 
            for i in iters:
                log_dir = "./logs/" + env + "." + alg + "." + str(steps) + "." + str(i)+"/"
                exists = os.path.isdir(log_dir)
                if exists: 
                    results_plotter.plot_results([log_dir], 1e7, results_plotter.X_TIMESTEPS, results_plotter.Y_REWARD, env+"--"+alg)
```
